```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R_ = 512 # known

H_ = 3 # global

h_ = 1.8 # person height
w_ = 0.4 # person width
```

```python
proj_x1 = np.linspace(0, R_, num=100)[1:-1]

real_x1 = 1 / np.sqrt((R_/proj_x1)**2 - 1) * H_
proj_x2 = (real_x1+w_) / np.sqrt((real_x1+w_)**2+(H_-h_)**2) * R_

y = proj_x2 - proj_x1

plt.figure(figsize=(12,4))
plt.plot(proj_x1, y)
plt.xlim(0,R_)
plt.ylim(0,300)
```

```python
r = np.linspace(0, 10, num=100)
x2 = (r+w_) / np.sqrt((r+w_)**2+(H_-h_)**2) * R_
x1 = r / np.sqrt(r**2 + H_**2) * R_

y = x2 - x1

plt.figure(figsize=(12,4))
plt.plot(r, y)
plt.xlim(0,10)
plt.ylim(0,300)
```
